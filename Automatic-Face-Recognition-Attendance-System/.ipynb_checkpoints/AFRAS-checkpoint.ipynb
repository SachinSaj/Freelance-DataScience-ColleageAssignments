```python
vid = cv2.VideoCapture(0)
fourcc = cv.VideoWriter_fourcc('m', 'p', '4', 'v')
out = cv2.VideoWriter(f'employee_{np.random.randint(10000)}.mp4', fourcc, 24, (1280, 720))
while(True):
    ret, frame = vid.read()
    out.write(frame) 
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release()
out.release() 
cv2.destroyAllWindows()
```
