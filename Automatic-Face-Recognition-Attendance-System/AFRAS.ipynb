```python

def convertToBinaryData(filename):
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData


student_name = input("Enter Your Name: ")
roll_no = input("Enter Your Roll Number")
vid = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter(f'Student_Face/{student_name}_{roll_no}.avi', fourcc, 10, (640, 480))
while(True):
    ret, frame = vid.read()
    out.write(frame) 
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
vid.release()
out.release() 
cv2.destroyAllWindows()

# Saving the data to database
sqliteConnection = sqlite3.connect('student_database.db')
cursor = sqliteConnection.cursor()
print("Connected to SQLite")
sqlite_insert_blob_query = """ INSERT INTO Student_data
                          (Roll, Name, Video) VALUES (?, ?, ?)"""
stuVideo = convertToBinaryData(f"Student_Face/{student_name}_{roll_no}.avi")
data_tuple = (roll_no, student_name, f"Student_Face/{student_name}_{roll_no}.avi")
cursor.execute(sqlite_insert_blob_query, data_tuple)
sqliteConnection.commit()
print("Student Video inserted successfully as a BLOB into a table")
cursor.close()
sqliteConnection.close()
print("the sqlite connection is closed")
```
